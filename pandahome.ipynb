```python
import numpy as np
import pandas as pd
movies = pd.read_table("C:\\Users\\shele\\Desktop\\pyton\\movielens\\movies.dat", delimiter = "::", 
                       names = ["MovieID","Title","Genres"], engine = "python")
rating = pd.read_table("C:\\Users\\shele\\Desktop\\pyton\\movielens\\ratings.dat", delimiter = "::", 
            names = ["UserID","MovieID","Ratings","Timestamp"], engine = "python")
users = pd.read_table("C:\\Users\\shele\\Desktop\\pyton\\movielens\\users.dat", delimiter = "::", 
            names = ["UserID","Gender","Age","Occupation","Zip-Code"], engine = "python")
rating = pd.merge(rating,users)
frame = pd.merge(movies,rating)
frame
```

```python
#Рейтинги по полу
rat = pd.pivot_table(frame, index = "Title", columns = "Gender", values = "Ratings", aggfunc='mean')
rat
```

```python
#Рейтинг во возрасту
rat1 = pd.pivot_table(frame, index = "Title", columns = "Age", values = "Ratings" )
rat1
```

```python
#Убывание популярности у женской аудитории
sort = rat.sort_values("F", ascending=False)
sort
```

```python
#Убывание популярности у возрастной аудитории

sort1 = rat1.sort_values(56, ascending=False)
sort1
```

```python
#Понравились женской аудитории
rat = pd.pivot_table(frame, index='Title', columns='Gender', values='Ratings', aggfunc='mean')
vs = rat["F"] - rat ["M"]
vs = vs.sort_values(ascending=False)
vs.head(15)
```
